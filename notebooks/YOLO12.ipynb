# Training Phase

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv12 model
model = YOLO('yolo12s.pt')

results = model.train(
    data='/content/dataset.yaml',
    epochs=100,
    imgsz=640,
    patience=20,
    batch=64,
    device='cuda',
    amp=True,
    workers=16,
    project="/content/drive/MyDrive/Deep Learning/YOLO12"
)
```

# Prediction Phase

```python
from ultralytics import YOLO

model = YOLO('train/weights/best.pt')

model.val(data='/content/dataset.yaml', split="test")
```

# Prediction on a single image

```python
from ultralytics import YOLO
import os, glob, cv2

# Where you saved training runs
project_dir = "/content/drive/MyDrive/Deep Learning/YOLO12"

# Try the default run folder, else pick the newest train*/weights/best.pt
best = f"{project_dir}/train/weights/best.pt"

# Load trained weights
model = YOLO(best)

# Your test image
img_path = "/content/ABOShipsDataset(YOLO)/images/test/201807041518_055.png"  # <- change this

# Quick visual: saves an annotated image
results = model.predict(
    source=img_path,
    conf=0.25,
    iou=0.5,
    imgsz=640,          # match your training size
    device=0,           # or 'cpu'
    save=True,
    project="/content",
    name="yolo12_infer"
)
print("Annotated image saved in: /content/yolo12_infer")
```
